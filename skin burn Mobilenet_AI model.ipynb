```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from keras.applications import MobileNet
from keras.models import Sequential
from keras.layers import GlobalAveragePooling2D, Dense
from matplotlib import pyplot as plt
from tensorflow.keras.optimizers import Adam
```

```python
image_folder = '/content/drive/MyDrive/microsoft_hackerground/data/'
class_folders = ['/content/scaled_1st degree burn',
                 '/content/scaled_2nd degree burn',
                 '/content/scaled_3nd degree burn']
classes = len(class_folders)
```

```python
images = []
labels = []
```

```python
for i, folder in enumerate(class_folders):
    folder_path = os.path.join(image_folder, folder)
    image_names = os.listdir(folder_path)
    for image_name in image_names:
        image = cv2.imread(os.path.join(folder_path, image_name))
        image = cv2.resize(image, (224, 224))  # 이미지 크기 조정
        images.append(image)
        labels.append(i)
```

```python
images = np.array(images)
labels = np.array(labels)
```

```python
train_images, test_images, train_labels, test_labels = train_test_split(images, labels, test_size=0.2, random_state=42)
```

```python
base_model = MobileNet(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
model = Sequential()
model.add(base_model)
model.add(GlobalAveragePooling2D())
model.add(Dense(64, activation='relu'))
model.add(Dense(128, activation='selu'))
model.add(Dense(classes, activation='softmax'))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit(train_images, train_labels, epochs=65, batch_size=32, validation_data=(test_images, test_labels))
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(accuracy))

# 정확도 그래프
plt.plot(epochs, accuracy, 'bo', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# 손실 그래프
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
validation_data = (test_images, test_labels)
evaluation = model.evaluate(test_images, test_labels)
print("Evaluation loss:", evaluation[0])
print("Evaluation accuracy:", evaluation[1])
```
